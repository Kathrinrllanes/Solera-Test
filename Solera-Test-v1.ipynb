# Step2 Detect the columns in which all their values are null¶

```python
#test
len(filtered_Dataset.columns)
```

```python
#Detect the columns in which all their values are null
nullColumns = []
numRows = filtered_Dataset.count()
for k in filtered_Dataset.columns:
    nullRows = filtered_Dataset.where(col(k).isNull()).count()#return the number of null rows for a column
    if nullRows ==  numRows: # i.e. if ALL values are NULL
        nullColumns.append(k)

nullColumns
```

```python
#drop null columns from the dataset
filtered_Dataset = filtered_Dataset\
.select([column for column in filtered_Dataset.columns if column not in nullColumns])
```

```python
#test
len(filtered_Dataset.columns)
```

# Step3 Detect the columns in which all their values are the same

```python
uniformColumns = []
numRows = filtered_Dataset.count()#the number of columns doesn't include the nullColumns
for k in filtered_Dataset.columns:
    # discard null values
    equalRows = filtered_Dataset.groupby(k).count().collect()[0][1]
    if equalRows ==  numRows: # i.e. the number of rows with the same values = total number of rows
        uniformColumns.append(k)

uniformColumns
```

```python
#eliminar esas columnas del dataset
#drop null columns from the dataset
filtered_Dataset = filtered_Dataset\
.select([column for column in filtered_Dataset.columns if column not in uniformColumns])
```

```python
#test
len(filtered_Dataset.columns)
```

```python
#Correlation between variables
#Correlation is a normalized measure of covariance.
#It provides quantitative measurements of the statistical dependence between two random variables.

#solo para varialbles numericas
from pyspark.sql.functions import rand
#filtered_Dataset.stat.corr('end_point_timestamp', 'end_point_timestamp_utc')
filtered_Dataset.stat.corr('top_speed', 'top_speed_mph')

#In the above example, top_speed correlates perfectly with top_speed_mph. 
```

```python
#Covariance between variables
filtered_Dataset.stat.cov('top_speed', 'top_speed_mph')
```

```python
#Cross Tabulation provides a table of the frequency distribution for a set of variables. 
#Cross-tabulation is a powerful tool in statistics that is used to observe the statistical significance 
#(or independence) of variables. 
filtered_Dataset.stat.crosstab('end_point_timestamp', 'end_point_timestamp_utc').show()
```
